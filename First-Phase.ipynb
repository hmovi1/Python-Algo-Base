Dicts and Sets

Note: freq is an empty dictionary. For keeping track of count

(Character x, the number of time it appears)

Why freq.get(c,0)?

For each iteration if there exists no element x, we count by default
0. This is important, otherwise it won't give desired count.

Whenever there's a frequency counting scenario this pattern appears to
be most useful.

```python
freq = {}

for c in "banana":
    freq[c] = freq.get(c,0) + 1

print(freq)    
```

Now we do the same but with Set.

Set = Unique Values

NO DUPLICATE ALLOWED

Set lookup = O(n) far better than any looped lookup.

Here seen.add(x) = Remember this particular entry for future checks,
since its already now an unique element of our constructed set.



```python
nums = [7, 2, 7, 5]

seen = set()

for x in nums:
    if x in seen:
        print(x)

    seen.add(x)
```
