# Sliding Window Intuition

## Why Expand?

Because we're exploring more data.

---

### Example: Longest Substring Without Repeating Characters

When we see:


````python
abc
````


everything is valid.

So we expand the window.

---

## Then Something Goes Wrong

Suppose:

```python
s = "abca"

Current window:

````python
a b c a
↑     ↑
l     r
````

Now the character a appears twice.

This violates the rule.

So the window becomes:

❌ INVALID

What Happens When Invalid?

This line activates:

````python
while invalid:
    l += 1
````

It is saying, I will shrink until the windows becomes valid again

# SHRINK Example

Current window:

````python
a b c a
↑     ↑
l     r
````

Duplicate character: a

Move l:

````python
a b c a
  ↑   ↑
  l   r
````

New window:

````python
b c a
````

Now the duplicate is gone.

Window is valid again.

Stop shrinking.

# This Is the Entire Sliding Window Idea

Expand

````python
r += 1
````

OR

````python
for r in range(len(s)):
````

Include more elements in the window.

Shrink

````python
l += 1
````

Remove elements from the left.

# Sliding Window Problem Conditions

#### Longest Substring Without Repeating Characters
### Condition:
- No duplicate characters in the window

---

#### Character Replacement
### Condition:
- At most k replacements allowed

---

#### Minimum Window Substring
### Condition:
- Window must contain all required characters

---

#### Maximum Sum Subarray
### Condition:
- Fixed window size k


# General Template

expand window

if rule is broken:
    shrink window

continue

# Key Insight

We never restart the process.

We never check all substrings.

That’s why Sliding Window is O(n) instead of O(n^2)

# The Mental Model

Imagine a rubber band:

l ------------ r


Expand:

r moves right → window grows

Shrink:

l moves right → window shrinks

## Goal:
Keep window valid while exploring all possibilities.

# Sliding Window Thinking Framework

## 1. What does my window represent?
Think about what your current window is holding.

Common cases:
- Substring (characters in a string)
- Subarray (elements in an array)
- Range of values (a continuous segment)

---

## 2. What makes the window invalid?
Identify the rule that breaks the window.

Common cases:
- Duplicate character appears
- Constraint is violated
- Required condition is not satisfied (missing elements / counts)

---

## 3. How do I restore validity?
Fix the window by shrinking from the left.

### Typical fix:
```python
l += 1

Continue shrinking until the window becomes valid again.

# Key Idea

Expand to explore, shrink to fix, repeat until done.
