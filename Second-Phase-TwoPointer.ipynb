How to move pointers?

The left pointer increases (Goes rightwards)
The right pointer decreases (Goes leftwards)

```python
l += 1
r -= 1
```

# Two Pointers Pattern Recognition Table

| Pattern | Condition Logic |
|----------|----------|
| Palindrome | `s[l] != s[r]` |
| Compare Ends | `arr[l] != arr[r]` |
| Reverse In Place | Swap values |
| Two Sum II | Compare sum to target |
| Container With Most Water | Move the shorter wall |
| Remove Duplicates | Compare current values |
| Merge Sorted Arrays | Take the smaller value |

# Two Pointers Cheat Sheet

| Pattern | What Are We Comparing? | Action |
|----------|----------|----------|
| Palindrome | Left char vs Right char | If different → fail |
| Compare Ends | Left value vs Right value | Check symmetry |
| Reverse In Place | Left element vs Right element | Swap |
| Two Sum II | `arr[l] + arr[r]` vs target | Move pointer to adjust sum |
| Container With Most Water | Left wall vs Right wall | Move shorter wall |
| Remove Duplicates | Current value vs previous unique value | Copy unique values |
| Merge Sorted Arrays | Current values from both arrays | Take smaller value |
